## ✅ Task 1.15: Run the samples above and figure out what returns.


###`merge()` — Complete Joins Guide


| Join Type | Keeps |
|---|---|
| Inner | Only matching rows |
| Left Outer | All left rows + matches |
| Right Outer | All right rows + matches |
| Full Outer | All rows from both |
| Left Anti | Left rows NOT in right |
| Right Anti | Right rows NOT in left |
| Full Anti | Rows unique to either side |

---

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import Circle
import warnings
warnings.filterwarnings('ignore')

# Sample DataFrames
df1 = pd.DataFrame({
    'id':   [1, 2, 3, 4],
    'name': ['Alice', 'Bob', 'Carol', 'Dave']
})

df2 = pd.DataFrame({
    'id':    [3, 4, 5, 6],
    'score': [90, 85, 78, 92]
})

print('df1 (left):'); print(df1.to_string(index=False))
print('\ndf2 (right):'); print(df2.to_string(index=False))
```

### 1) Inner Join
Returns **only rows whose key exists in both** DataFrames. This is the **default** behaviour of `pd.merge()`.

```python
inner = pd.merge(df1, df2, on='id')  # how='inner' is default
print('Inner Join:'); print(inner)
```

### 2) Left Outer Join
Returns **all rows from the left** DataFrame, filling `NaN` where the right has no match.

```python
left_join = pd.merge(df1, df2, on='id', how='left')
print('Left Outer Join:'); print(left_join)
```

### 3) Right Outer Join
Returns **all rows from the right** DataFrame, filling `NaN` where the left has no match.

```python
right_join = pd.merge(df1, df2, on='id', how='right')
print('Right Outer Join:'); print(right_join)
```

### 4) Full Outer Join
Returns **all rows from both** DataFrames, using `NaN` wherever there is no match.

```python
full_outer = pd.merge(df1, df2, on='id', how='outer')
print('Full Outer Join:'); print(full_outer)
```

### 5) Left Anti-Join
Returns rows **only in the left** DataFrame — keys that have **no match** in the right.

> **Two-step approach:** left-join with `indicator=True`, then filter for `'left_only'` rows.

```python
df3 = pd.merge(df1, df2, on='id', how='left', indicator=True)
print('Step 1 — with _merge indicator:'); print(df3)

left_anti_ids = df3.loc[df3['_merge'] == 'left_only', 'id']
result = df1[df1['id'].isin(left_anti_ids)]
print('\nLeft Anti-Join:'); print(result)
```

### 6) Right Anti-Join
Returns rows **only in the right** DataFrame. Same approach — filter for `'right_only'`.

```python
df3 = pd.merge(df1, df2, on='id', how='right', indicator=True)
right_anti_ids = df3.loc[df3['_merge'] == 'right_only', 'id']
result = df2[df2['id'].isin(right_anti_ids)]
print('Right Anti-Join:'); print(result)
```

### 7) Full Anti-Join (Symmetric Difference)
Rows that are **unique to either** DataFrame — everything NOT shared.

```python
df3 = pd.merge(df1, df2, on='id', how='outer', indicator=True)
result = df3[df3['_merge'] != 'both'][['id', '_merge']]
print('Full Anti-Join:'); print(result)
```

### 8) Merging on Different Column Names
Use `left_on` and `right_on` when key columns have **different names** in each DataFrame.

```python
df_a = pd.DataFrame({'student_id': [1, 2, 3], 'name':  ['Alice', 'Bob', 'Carol']})
df_b = pd.DataFrame({'sid':         [2, 3, 4], 'grade': ['A', 'B', 'C']})

merged = pd.merge(df_a, df_b, left_on='student_id', right_on='sid')
print('Merged on different column names:'); print(merged)
```

### 9) Merging on Multiple Columns
Pass a **list** to `on` to merge on two or more columns simultaneously.

```python
df_x = pd.DataFrame({'A':[1,1,2],'B':['x','y','x'],'val1':[10,20,30]})
df_y = pd.DataFrame({'A':[1,2,2],'B':['x','x','y'],'val2':[100,200,300]})

multi = pd.merge(df_x, df_y, on=['A', 'B'])
print('Merged on multiple columns:'); print(multi)
```
